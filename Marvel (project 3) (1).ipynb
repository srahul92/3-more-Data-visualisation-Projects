# Data Visualisation on marvel characters

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
marvel=pd.read_csv('charcters_stats.csv')
```

```python
marvel.head()
```

```python
marvel.shape
```

```python
marvel['Alignment'].value_counts()
```

```python
marvel[marvel['Alignment']=='good']
```

```python
good=marvel[marvel['Alignment']=='good']
```

```python
good.head()
```

```python
good.sort_values(by=['Speed'],ascending=False).head()
```

```python
good.sort_values(['Power'],ascending=False).head()
```

```python
good_max_power=good[good['Power']==100]
```

```python
good_max_power.head()
```

```python
good_max_power.shape
```

```python
good_max= good.sort_values(['Total'],ascending=False).head()
```

```python
good_max.head()
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(good_max['Name'])[0:5],list(good_max['Total'])[0:5],color='g')
plt.show()

```
